```python
import pandas as pd
import os
import re
import atrader as at
import numpy as np
import matplotlib.pyplot as plt
```

```python
#获取上证A股代码
dirs = os.listdir("./data")
def get_name(s):
    return re.findall('([0123456789]{1,}).',s)
target_list = []
for i in dirs:
    s = get_name(i)
    if len(s)!= 0:
        target = 'SSE.'+s[0]
        target_list.append(target)
for file_name in target_list:
    data = pd.read_csv('./data/'+file_name+'.csv',index_col=0)
    data['time'] = pd.to_datetime(data['time']).astype("datetime64[D]")
    data.set_index(data['time'],inplace=True)
    if int(str(data['time'][-1])[:4])-2021 < 0:
        target_list.remove(file_name)
```

```python
#加载数据
def load_data(file_path,fn):
    data = pd.read_csv(file_path,index_col=0)
    data['time'] = pd.to_datetime(data['time']).astype("datetime64[D]")
    #del data['Unnamed: 0']
    del data['open_interest']
    data.set_index(data['time'],inplace=True)
    #del data['time']
    yz = at.get_factor_by_code(factor_list=['PE','PB','PCF','PS','MKV','ARBR','BP'], target=fn, begin_date=data['time'][0], end_date=data['time'][-1])
    yz.set_index(yz['date'],inplace=True)
    yz.drop(index = list(set(yz['date'])-set(data['time'])))
    data['PE'] = yz['PE']
    data['PB'] = yz['PB']
    data['PCF'] = yz['PCF']
    data['PS'] = yz['PS']
    data['MKV'] = yz['MKV']
    data['ARBR'] = yz['ARBR']
    data['BP'] = yz['BP']
    data['gains'] = data['close']-data['open']
    data['volatility'] = (data['high']-data['low'])
    data['ER'] = data['amount']/data['volume']
    return data
```

```python
data = load_data('./data/SSE.600089.csv','SSE.600089')
```
